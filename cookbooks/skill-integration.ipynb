# 在 Claude / Cursor / Codex 中接入 Sciverse Skill

> 安装 Sciverse MCP 工具，让 AI 助手直接调用科学文献检索

**场景**: 开发者希望在日常使用的 AI 编程助手（Claude Code、Cursor、Codex CLI）中直接调用 Sciverse 检索科学文献，无需手动写 API 调用代码。

**使用接口**: Sciverse MCP Server, semantic_search, read_content, list_catalog

**预估调用量**: ~2–5 次工具调用 / 一次对话

---


## Step 1: 安装 Sciverse Skill

根据你使用的 AI 助手选择对应的安装方式


````python
# ─── Claude Code ───
# 方式 A：通过官方域名安装（推荐）
!npx skills add https://sciverse.space
# 方式 B：从 GitHub 源安装
!npx skills add opendatalab/Sciverse-Agent-Tools --skill sciverse
# ─── Cursor ───
# 在 Cursor Settings > MCP 中添加 Sciverse MCP Server
# Server URL: https://mcp.sciverse.space
# ─── Codex CLI ───
# 参考 Codex 文档配置 MCP server
````

## Step 2: 配置 API Token

设置环境变量，Skill 会自动读取


````python
# 在 shell 配置文件中添加（~/.bashrc 或 ~/.zshrc）
import os
os.environ["SCIVERSE_API_TOKEN"] = "sv-your-token-here"  # 替换为你的真实值
# 或在项目 .env 文件中
!SCIVERSE_API_TOKEN=sv-your-token-here
# 验证安装成功（Claude Code）
!npx skills list | grep sciverse
````

## Step 3: 在 AI 助手中使用

安装后直接在对话中提问，AI 会自动调用 Sciverse 工具


```markdown
# 在 Claude Code / Cursor / Codex 中直接提问：

> 帮我查找 2023 年以来关于 LLM 幻觉检测的论文

AI 助手会自动：
1. 调用 semantic_search(query="LLM hallucination detection", top_k=10)
2. 返回相关论文片段和引用
3. 如需详情，继续调用 read_content 获取全文

> 用结构化检索按 Nature 期刊过滤

AI 助手会：
1. 调用 list_catalog() 确认字段名
2. 调用 search_papers(filters=[...]) 执行过滤
```


## 注意事项

- SCIVERSE_API_TOKEN 环境变量必须设置，否则所有工具调用会返回 401
- 不同 AI 助手的 MCP/Skill 配置方式不同，请参考各自官方文档
- Skill 安装后对支持 MCP 协议的 AI 助手生效
- 具体调用限制请在 Token 管理页查看你的配额

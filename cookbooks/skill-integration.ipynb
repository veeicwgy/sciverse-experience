3. 将下方 `sv-...` 替换为你的真实 Token


```python
# 安装依赖（Colab 环境）
!pip install -q httpx anthropic
```

## Step 1: 安装 Sciverse Skill

在支持 MCP 的 Agent 平台中安装 Sciverse 工具


```python
# 在 Claude Desktop / Cursor 的 MCP 配置中添加：
# ~/.config/claude/mcp_servers.json (或对应平台配置文件)

config = {
    "sciverse": {
        "command": "npx",
        "args": ["-y", "@sciverse/mcp-server"],
        "env": {
            "SCIVERSE_API_KEY": "sv-..."
        }
    }
}

# 安装完成后，Agent 将自动获得以下工具：
# - sciverse_search: 语义检索科学文献
# - sciverse_read: 读取论文全文
# - sciverse_meta: 结构化元数据查询
# - sciverse_resource: 下载论文附件

print("Sciverse Skill 配置完成")
print("Agent 现在可以直接调用科学文献检索能力")
```

## Step 2: 在 Agent 中使用

Agent 会自动识别何时需要调用 Sciverse


```python
# Agent 对话示例：
# 用户: "帮我找 AlphaFold2 的核心创新点，引用原文"
#
# Agent 内部调用链路:
# 1. sciverse_search(query="AlphaFold2 core innovations architecture")
# 2. sciverse_read(doc_id="af2_nature_2021", offset=0)
# 3. 整合证据，生成带引用的回答

# 验证 Skill 是否正常工作
import subprocess
result = subprocess.run(
    ["npx", "-y", "@sciverse/mcp-server", "--test"],
    capture_output=True, text=True
)
print(result.stdout)
```
